2. column 이름 따서 con_(region1)_(region2)_count 정보를 추출

3. region 1, region 2를 행과 열 이름으로 넣기 - 필요한가?

4. con_region 1_region 2의 값을 [region 1][region 2] 위치에 넣기

```python
sample=dti.loc[0:0]
```

```python
collist=list(dti.columns)[1:]
```

```python
collist
```

```python
ROI1=[]
ROI2=[]
for s in collist:
    ROI1_element=s.split('_')[1]
    ROI1.append(ROI1_element)
    ROI2_element=s.split('_')[2]
    ROI2.append(ROI2_element)    
```

```python
ROI=ROI1+ROI2
```

```python
len(set(ROI))
```

```python
adjmat=pd.DataFrame(data=np.zeros((84,84)), index=list(set(ROI)), columns=list(set(ROI)))
```

```python
adjmat
```

```python
for s in collist:
    a=s.split('_')[1]
    b=s.split('_')[2]
    adjmat[a][b]=sample[s].tolist()[0]
    adjmat[b][a]=sample[s].tolist()[0]
        
for i in set(ROI):
    for j in set(ROI):
        if i == j:
            adjmat[i][j] = 1       
```

```python
adjmat
```

```python
for k in range (dti.shape[0]):
    sample=data.loc[k:k]
    conmat=pd.DataFrame(data=np.zeros((84,84)), index=list(set(ROI)), columns=list(set(ROI))) #초기화
    subject=sample['subjectkey'].tolist()[0]
    for s in collist:
        a=s.split('_')[1]
        b=s.split('_')[2]
        conmat[a][b]=sample[s].tolist()[0]
        conmat[b][a]=sample[s].tolist()[0]

    for i in set(ROI):
        for j in set(ROI):
            if i == j:
                conmat[i][j] = 1
    conmat.to_csv("/home/ubuntu/Stella/MLVU multimodality/DTI/DTI_con_mat_1026/DTI connectivity matrix %s.csv" %subject, header=True, index=True)
```
