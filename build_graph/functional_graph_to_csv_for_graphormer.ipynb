### 3-1 graph level - graph label, num_edge_list, num_node_list

```python
graph_label = pd.DataFrame(columns=['intelligence'])
num_edge = pd.DataFrame(columns=['num_edge'])
num_node = pd.DataFrame(columns=['num_node'])
```

```python
type(test['num_nodes'])
```

```python
graph_list[0].split('_')[2].split('.')[0]
```

```python
for i in range(len(graph_list)):
    data = os.path.join(data_dir, graph_list[i])
    subject_id = graph_list[i].split('_')[2].split('.')[0]
    with open(data, 'rb') as f:
        graph_data = pickle.load(f)
        
    # 1 graph label
    intelligence_score = np.float16(graph_data['labels'])
    graph_label = graph_label.append(pd.DataFrame([intelligence_score], columns=['intelligence']),
                                     ignore_index=True)
    
    
    # 2 num_edge_list
    num_edge_score = int(graph_data['num_edges'])
    num_edge = num_edge.append(pd.DataFrame([num_edge_score], columns=['num_edge']),
                               ignore_index=True)
    
    
    # 3 num_node_list
    num_node_score = int(graph_data['num_nodes'])
    num_node = num_node.append(pd.DataFrame([num_node_score], columns=['num_node']),
                               ignore_index=True)
```

```python
# save files    
graph_label.to_csv('graph-label.csv', index=False, header=None)
num_edge.to_csv('num-edge-list.csv', index=False, header=None)
num_node.to_csv('num-node-list.csv', index=False, header=None)
```

### 3-2 edge level - edge_feat, edge

```python
edge_feat_list = []
edge_index_source_list=[]
edge_index_target_list=[]
```

```python
test['edge_index'][0][0]
```

```python
type(round(graph_data['edge_index'][0][0]))
```

```python
for i in range(len(graph_list)):
    data = os.path.join(data_dir, graph_list[i])
    subject_id = graph_list[i].split('_')[2].split('.')[0]
    with open(data, 'rb') as f:
        graph_data = pickle.load(f)
    
    for j in range(graph_data['num_edges']):
        # 4 edge_feat
        edge_feat_list.append(int(10*graph_data['edge_feat'][j].astype(np.float64)))
        
        # 5 edge_index
        edge_index_source_list.append(round(graph_data['edge_index'][0][j]))
        edge_index_target_list.append(round(graph_data['edge_index'][1][j]))
```

```python
# save edge feat
edge_feat=pd.DataFrame({'struct_conn': edge_feat_list})
edge_feat.to_csv('edge-feat.csv', index=False, header=None)

# save edge index
edge_index=pd.DataFrame({'source': edge_index_source_list, 'target' : edge_index_target_list})
edge_index.to_csv('edge.csv', index=False, header=None)
```

```python
for i in range(len(graph_list)):
    data = os.path.join(data_dir, graph_list[i])
    subject_id = graph_list[i].split('_')[2].split('.')[0]
    with open(data, 'rb') as f:
        graph_data = pickle.load(f)
    
    for j in range(graph_data['num_nodes']):
        # 6 node_feat
        ## 6-1 cortical thickness
        if str(graph_data['node_feat'].T[0][j]) == 'nan':
            node_feat_cor_thick_list.append(-1)
        else:
            node_feat_cor_thick_list.append(int(10 * graph_data['node_feat'].T[0][j].astype(np.float32)))
        
        ## 6-2 volume - scaled
        if str(graph_data['node_feat'].T[1][j]) == 'nan':
            node_feat_vol_list.append(-1)
        else:
            node_feat_vol_list.append(int(0.001 * graph_data['node_feat'].T[1][j].astype(np.float32)))
```
